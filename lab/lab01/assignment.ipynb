## Assignment: Linear Regression and MSE

This assignment contains two questions:

1. Calculate the coefficients of a linear regression model.
2. Calculate the Mean Squared Error (MSE) between two arrays.

### Question 1: Linear Regression

Fit a linear regression model using the given dataset. Store the coefficients in a variable named `coefficients`.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])  # Features
y = np.dot(X, np.array([1.5, 2.0])) + 3         # Labels

# Write your solution below:
model = LinearRegression()
model.fit(X, y)
coefficients = model.coef_  # Store coefficients
```

### Question 2: Mean Squared Error (MSE)

Calculate the Mean Squared Error (MSE) between two arrays: `y_true` and `y_pred`. Store the result in a variable named `mse`.

```python
# Example Arrays
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

# Write your solution below:
mse = np.mean((y_true - y_pred) ** 2)  # Calculate Mean Squared Error
```
